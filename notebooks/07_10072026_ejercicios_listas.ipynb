```python
#Ejercicios01 
numeros = [4, 7, 2, 4, 9, 7, 1, 2, 10, 4]

numeros_distinct = []

for numero in numeros:
    if numero not in numeros_distinct:
        numeros_distinct.append(numero)

print(numeros_distinct)
```

```python
# Ejercicio02
numeros = [12, 7, 9, 24, 31, 18, 5, 42, 13, 8]

numeros_pares = [
    numero
    for numero in numeros
    if numero % 2 == 0
]

numeros_impares = [
    numero
    for numero in numeros
    if numero % 2 == 1
]

print(f"numeros pares: {numeros_pares}")
print(f"numeros impares: {numeros_impares}")
```

```python
# Ejercicio04

palabras = ['python', 'sol', 'programacion', 'lista', 'computadora', 'if', 'variable']

contador = 0

for palabra in palabras:
    if len(palabra) > 5:
        contador += 1

print("Cantidad de palabras con más de 5 letras:", contador)
```

```python
# Ejercicio 05

notas = [14, 18, 11, 16, 9, 13, 20]

suma = 0

for nota in notas:
    suma += nota

promedio = suma / len(notas)

print("Promedio:", promedio)

if promedio >= 11:
    print("Aprobó")
else:
    print("Desaprobó")
```

```python
# Ejercicio09

lista1 = [4, 8, 12, 16, 20]
lista2 = [8, 10, 12, 18, 20]

lista_unicos = []

# Buscar los elementos únicos de la primera lista
for numero in lista1:
    if numero not in lista2:
        lista_unicos.append(numero)

# Buscar los elementos únicos de la segunda lista
for numero in lista2:
    if numero not in lista1:
        lista_unicos.append(numero)

print(lista_unicos)
```

```python
# Ejercicio 20

estudiantes = ['Ana', 'Luis', 'Carlos', 'María', 'Pedro']
notas = [18, 10, 14, 8, 16]

aprobados = []

for estudiante, nota in zip(estudiantes, notas):
    if nota >= 11:
        aprobados.append(estudiante)

print(aprobados)
```
