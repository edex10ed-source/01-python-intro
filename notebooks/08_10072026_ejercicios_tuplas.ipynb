```python
# Ejercicio 01
numeros = (15, 8, 42, 3, 27, 19, 5)

# Obtener el mayor y el menor
mayor = max(numeros)
menor = min(numeros)

# Calcular la diferencia
diferencia = mayor - menor

# Mostrar resultados
print("Número mayor:", mayor)
print("Número menor:", menor)
print("Diferencia:", diferencia)
```

```python
# Ejercicio 04
notas = (18, 15, 20, 11, 14, 17, 19)

# Inicializamos las variables
mayor = notas[0]      
menor = notas[0]      
suma = 0              
aprobados = 0         

# Recorremos la tupla
for nota in notas:

    # Sumamos cada nota
    suma = suma + nota

    # Buscamos la nota mayor
    if nota > mayor:
        mayor = nota

    # Buscamos la nota menor
    if nota < menor:
        menor = nota

    # Contamos las notas aprobadas (11 o más)
    if nota >= 11:
        aprobados = aprobados + 1

# Calculamos el promedio
promedio = round(suma /len(notas),2)

# Mostramos los resultados
print("Promedio:",promedio)                      
print("Nota máxima:", mayor)
print("Nota mínima:", menor)
print("Cantidad de aprobados:", aprobados)
```
