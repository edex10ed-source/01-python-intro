```python
# Ejercicio 01
numeros = (15, 8, 42, 3, 27, 19, 5)

# Obtener el mayor y el menor
mayor = max(numeros)
menor = min(numeros)

# Calcular la diferencia
diferencia = mayor - menor

# Mostrar resultados
print("Número mayor:", mayor)
print("Número menor:", menor)
print("Diferencia:", diferencia)
```

```python
#Ejercicio 02

datos = (5, 8, 5, 1, 9, 5, 2, 8, 5, 3)

#Ingrese numero 
numero = int(input("Ingrese un número: "))
#Contar numnero 
cantidad = datos.count(numero)
#Mostrar resultados:

print("El número", numero, "aparece", cantidad, "veces.")
```
