```python
#Ejercicio 01
estudiantes_python = {
    "Ana", "Luis", "María", "Carlos", "Sofía", "Diego"
}

estudiantes_javascript = {
    "María", "Carlos", "Pedro", "Lucía", "Diego", "Valeria"
}

# 1. Estudiantes inscritos en ambos cursos (Intersección)
ambos = estudiantes_python.intersection(estudiantes_javascript)
print("Estudiantes en ambos cursos:", ambos)

# 2. Estudiantes que solo están en Python (Diferencia)
solo_python = estudiantes_python.difference(estudiantes_javascript)
print("Solo en Python:", solo_python)

# 3. Estudiantes que solo están en JavaScript (Diferencia)
solo_javascript = estudiantes_javascript.difference(estudiantes_python)
print("Solo en JavaScript:", solo_javascript)

# 4. Todos los estudiantes sin repetir nombres (Unión)
todos = estudiantes_python.union(estudiantes_javascript)
print("Todos los estudiantes:", todos)
```

```python
#Ejercicios 02
almacen_arequipa = {
    "laptop", "teclado", "mouse", "monitor", "impresora", "audifonos"
}

almacen_lima = {
    "laptop", "mouse", "monitor", "tablet", "webcam", "parlantes"
}

# 1. Productos disponibles en ambos almacenes (Intersección)
ambos = almacen_arequipa.intersection(almacen_lima)
print("Productos en ambos almacenes:", ambos)

# 2. Productos exclusivos de Arequipa (Diferencia)
solo_arequipa = almacen_arequipa.difference(almacen_lima)
print("Solo en Arequipa:", solo_arequipa)

# 3. Productos exclusivos de Lima (Diferencia)
solo_lima = almacen_lima.difference(almacen_arequipa)
print("Solo en Lima:", solo_lima)

# 4. Todos los productos sin repetir (Unión)
todos = almacen_arequipa.union(almacen_lima)
print("Todos los productos:", todos)

# 5. ¿Tienen exactamente el mismo inventario?
print("¿Tienen el mismo inventario?", almacen_arequipa == almacen_lima)
```

```python
# Ejercicio 03
#Permisos usuarios 
permisos_usuario = {
    "leer", "editar", "comentar", "descargar"
}

# Permisos requeridos
permisos_requeridos = {
    "leer", "editar", "eliminar", "publicar"
}

# 1. ¿El usuario posee todos los permisos requeridos?git
print("¿Tiene todos los permisos requeridos?")
print(permisos_usuario.issuperset(permisos_requeridos))

# 2. Permisos que le faltan
faltan = permisos_requeridos.difference(permisos_usuario)
print("\nPermisos que le faltan:")
print(faltan)

# 3. Permisos adicionales que tiene
adicionales = permisos_usuario.difference(permisos_requeridos)
print("\nPermisos adicionales:")
print(adicionales)

# 4. Cantidad total de permisos diferentes
todos = permisos_usuario.union(permisos_requeridos)
print("\nCantidad total de permisos diferentes:")
print(len(todos))
```

```python
#Ejercicios 04
numeros_a = {2, 4, 6, 8, 10, 12, 14}
numeros_b = {4, 6, 9, 10, 13, 14, 16}
numeros_c = {1, 4, 6, 10, 11, 14, 18}

# 1. Números presentes en los tres conjuntos
tres_conjuntos = numeros_a.intersection(numeros_b).intersection(numeros_c)
print("En los tres conjuntos:", tres_conjuntos)

# 2. Números presentes en A y B, pero no en C
a_b = numeros_a.intersection(numeros_b)
resultado = a_b.difference(numeros_c)
print("En A y B, pero no en C:", resultado)

# 3. Números que solo pertenecen a A
solo_a = numeros_a.difference(numeros_b).difference(numeros_c)
print("Solo en A:", solo_a)

# 4. Todos los números diferentes
todos = numeros_a.union(numeros_b).union(numeros_c)
print("Todos los números diferentes:", todos)
```

```python
# Ejercicio 05
habilidades_candidato_1 = {
    "Python", "Git", "SQL", "Docker", "React"
}

habilidades_candidato_2 = {
    "Python", "JavaScript", "React", "Node.js", "Git"
}

habilidades_requeridas = {
    "Python", "Git", "SQL", "Docker"
}

# 1. Habilidades que comparten ambos candidatos
comparten = habilidades_candidato_1.intersection(habilidades_candidato_2)
print("Habilidades en común:", comparten)

# 2. Habilidades exclusivas de cada candidato
solo_candidato_1 = habilidades_candidato_1.difference(habilidades_candidato_2)
solo_candidato_2 = habilidades_candidato_2.difference(habilidades_candidato_1)

print("Solo candidato 1:", solo_candidato_1)
print("Solo candidato 2:", solo_candidato_2)

# 3. Habilidades requeridas que le faltan a cada candidato
faltan_candidato_1 = habilidades_requeridas.difference(habilidades_candidato_1)
faltan_candidato_2 = habilidades_requeridas.difference(habilidades_candidato_2)

print("Le faltan al candidato 1:", faltan_candidato_1)
print("Le faltan al candidato 2:", faltan_candidato_2)

# 4. ¿Qué candidato cumple con más requisitos?
cumple_candidato_1 = habilidades_candidato_1.intersection(habilidades_requeridas)
cumple_candidato_2 = habilidades_candidato_2.intersection(habilidades_requeridas)

if len(cumple_candidato_1) > len(cumple_candidato_2):
    print("El candidato 1 cumple con más requisitos.")
elif len(cumple_candidato_2) > len(cumple_candidato_1):
    print("El candidato 2 cumple con más requisitos.")
else:
    print("Ambos cumplen la misma cantidad de requisitos.")

# 5. Todas las habilidades sin repetir
todas = habilidades_candidato_1.union(habilidades_candidato_2)
print("Todas las habilidades:", todas)
```
