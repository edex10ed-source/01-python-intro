```python
#Ejercicio 01
estudiantes_python = {
    "Ana", "Luis", "María", "Carlos", "Sofía", "Diego"
}

estudiantes_javascript = {
    "María", "Carlos", "Pedro", "Lucía", "Diego", "Valeria"
}

# 1. Estudiantes inscritos en ambos cursos (Intersección)
ambos = estudiantes_python.intersection(estudiantes_javascript)
print("Estudiantes en ambos cursos:", ambos)

# 2. Estudiantes que solo están en Python (Diferencia)
solo_python = estudiantes_python.difference(estudiantes_javascript)
print("Solo en Python:", solo_python)

# 3. Estudiantes que solo están en JavaScript (Diferencia)
solo_javascript = estudiantes_javascript.difference(estudiantes_python)
print("Solo en JavaScript:", solo_javascript)

# 4. Todos los estudiantes sin repetir nombres (Unión)
todos = estudiantes_python.union(estudiantes_javascript)
print("Todos los estudiantes:", todos)
```

```python
#Ejercicios 02
almacen_arequipa = {
    "laptop", "teclado", "mouse", "monitor", "impresora", "audifonos"
}

almacen_lima = {
    "laptop", "mouse", "monitor", "tablet", "webcam", "parlantes"
}

# 1. Productos disponibles en ambos almacenes (Intersección)
ambos = almacen_arequipa.intersection(almacen_lima)
print("Productos en ambos almacenes:", ambos)

# 2. Productos exclusivos de Arequipa (Diferencia)
solo_arequipa = almacen_arequipa.difference(almacen_lima)
print("Solo en Arequipa:", solo_arequipa)

# 3. Productos exclusivos de Lima (Diferencia)
solo_lima = almacen_lima.difference(almacen_arequipa)
print("Solo en Lima:", solo_lima)

# 4. Todos los productos sin repetir (Unión)
todos = almacen_arequipa.union(almacen_lima)
print("Todos los productos:", todos)

# 5. ¿Tienen exactamente el mismo inventario?
print("¿Tienen el mismo inventario?", almacen_arequipa == almacen_lima)
```

```python
# Ejercicio 03
permisos_usuario = {
    "leer", "editar", "comentar", "descargar"
}

# Permisos requeridos
permisos_requeridos = {
    "leer", "editar", "eliminar", "publicar"
}

# 1. ¿El usuario posee todos los permisos requeridos?
print("¿Tiene todos los permisos requeridos?")
print(permisos_usuario.issuperset(permisos_requeridos))

# 2. Permisos que le faltan
faltan = permisos_requeridos.difference(permisos_usuario)
print("\nPermisos que le faltan:")
print(faltan)

# 3. Permisos adicionales que tiene
adicionales = permisos_usuario.difference(permisos_requeridos)
print("\nPermisos adicionales:")
print(adicionales)

# 4. Cantidad total de permisos diferentes
todos = permisos_usuario.union(permisos_requeridos)
print("\nCantidad total de permisos diferentes:")
print(len(todos))
```
