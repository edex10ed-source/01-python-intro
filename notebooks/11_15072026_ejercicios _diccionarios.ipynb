```python
# Ejercicio 1

inventario_norte = {"laptop": 4, "mouse": 10, "monitor": 3}
inventario_sur = {"laptop": 2, "teclado": 8, "monitor": 5}
inventario_centro = {"mouse": 6, "teclado": 4, "webcam": 7}

# Copia del inventario norte
inventario_total = dict(inventario_norte)

# Agregar inventario sur
for llave, valor in inventario_sur.items():

    if llave not in inventario_total:
        inventario_total[llave] = 0

    inventario_total[llave] += valor

# Agregar inventario centro
for llave, valor in inventario_centro.items():

    if llave not in inventario_total:
        inventario_total[llave] = 0

    inventario_total[llave] += valor

# Ordenar las llaves
llaves_ordenadas = sorted(inventario_total.keys())

# Crear un nuevo diccionario ordenado
inventario_total_con_llaves_ordenadas = {}

for llave in llaves_ordenadas:
    inventario_total_con_llaves_ordenadas[llave] = inventario_total[llave]

# Mostrar resultados
print(inventario_total)
print(inventario_total_con_llaves_ordenadas)
```

```python

# Ejercicio 2

calificaciones = {
    "Ana": [17, 18, 16, 19],
    "Luis": [15, 16, 17, 16],
    "Carla": [18, 17, 19, 18],
    "Pedro": [16, 16, 16, 16],
    "Lucía": [15, 17, 16, 16]
}

# Diccionario para guardar los resultados
resultado = {}

# Calcular promedio, máxima y mínima
for estudiante, notas in calificaciones.items():

    promedio = sum(notas) / len(notas)
    maxima = max(notas)
    minima = min(notas)

    resultado[estudiante] = {
        "promedio": promedio,
        "nota_maxima": maxima,
        "nota_minima": minima
    }

# Crear el ranking
ranking = sorted(
    resultado.items(),
    key=lambda x: (-x[1]["promedio"], x[0])
)

# Mostrar resultados
print("Información de los estudiantes")
for estudiante, datos in resultado.items():
    print(estudiante, "->", datos)

print("\nRanking")

for puesto, (estudiante, datos) in enumerate(ranking, start=1):
    print(
        puesto,
        estudiante,
        "Promedio:", datos["promedio"],
        "Máxima:", datos["nota_maxima"],
        "Mínima:", datos["nota_minima"]
    )
```

```python
#Ejercicios 5
empleados = [
    {"nombre": "Ana", "departamento": "Desarrollo", "nivel": "Senior", "salario": 5200},
    {"nombre": "Luis", "departamento": "Diseño", "nivel": "Junior", "salario": 2800},
    {"nombre": "Carla", "departamento": "Desarrollo", "nivel": "Junior", "salario": 3400},
    {"nombre": "Pedro", "departamento": "Desarrollo", "nivel": "Senior", "salario": 5600},
    {"nombre": "Lucía", "departamento": "Diseño", "nivel": "Senior", "salario": 4700}
]

grupos = {}

for empleado in empleados:

    departamento = empleado["departamento"]
    nivel = empleado["nivel"]

    grupos.setdefault(departamento, {})
    grupos[departamento].setdefault(
        nivel,
        {
            "nombres": [],
            "cantidad": 0,
            "suma_salarios": 0
        }
    )

    grupos[departamento][nivel]["nombres"].append(empleado["nombre"])
    grupos[departamento][nivel]["cantidad"] += 1
    grupos[departamento][nivel]["suma_salarios"] += empleado["salario"]

for departamento in grupos:

    for nivel in grupos[departamento]:

        datos = grupos[departamento][nivel]

        datos["salario_promedio"] = datos["suma_salarios"] / datos["cantidad"]

        del datos["suma_salarios"]


print("AGRUPACIÓN DE EMPLEADOS\n")

for departamento in grupos:

    print("Departamento:", departamento)

    for nivel in grupos[departamento]:

        print("  Nivel:", nivel)
        print("    Nombres:", grupos[departamento][nivel]["nombres"])
        print("    Cantidad:", grupos[departamento][nivel]["cantidad"])
        print("    Salario promedio:", grupos[departamento][nivel]["salario_promedio"])

    print()
```

```python
#Ejercicios 6
tienda_a = {
    "laptop": 3500,
    "mouse": 80,
    "teclado": 160,
    "monitor": 920
}

tienda_b = {
    "laptop": 3400,
    "mouse": 90,
    "teclado": 150,
    "webcam": 220
}

productos = list(tienda_a.keys()) + list(tienda_b.keys())
productos = set(productos)

comparativo = {}

for producto in productos:

    if producto in tienda_a and producto in tienda_b:

        if tienda_a[producto] > tienda_b[producto]:
            mas_barato = "tienda_b"
            precio = tienda_b[producto]

        elif tienda_a[producto] == tienda_b[producto]:
            mas_barato = "es igual en ambas tiendas"
            precio = tienda_a[producto]

        else:
            mas_barato = "tienda_a"
            precio = tienda_a[producto]

        comparativo[producto] = {
            "diferencia": abs(tienda_a[producto] - tienda_b[producto]),
            "mas_barato": mas_barato,
            "precio": precio
        }

    if producto in tienda_a and producto not in tienda_b:

        comparativo[producto] = {
            "diferencia": "no hay diferencia",
            "mas_barato": "tienda_a",
            "precio": tienda_a[producto]
        }

    if producto in tienda_b and producto not in tienda_a:

        comparativo[producto] = {
            "diferencia": "no hay diferencia",
            "mas_barato": "tienda_b",
            "precio": tienda_b[producto]
        }

for comp in comparativo:
    print(comp)
    print(comparativo[comp])
    print("-" * 100)
```

```python
# Ejercicio 7

candidatos = {"Ana", "Luis", "Carla"}

votos = ["Ana", "Luis", "ana", "Carla", "Pedro", "Luis", "ANA", "", "Carla"]

mapa = {
    "ana": "Ana",
    "luis": "Luis",
    "carla": "Carla"
}

conteo = {}

for voto in votos:

    voto_limpio = voto.lower()

    candidato = mapa.get(voto_limpio, "invalido")

    if candidato == "invalido":
        conteo.setdefault("Votos invalidos", 0)
        conteo["Votos invalidos"] += 1
        continue

    conteo.setdefault(candidato, 0)
    conteo[candidato] += 1

total_votos = len(votos)

for candidato in candidatos:
    conteo[f"{candidato}_pct"] = round(
        conteo[candidato] / total_votos * 100,
        2
    )

conteo["Votos invalidos_pct"] = round(
    conteo["Votos invalidos"] / total_votos * 100,
    2
)

mayor = 0

for candidato in candidatos:
    if conteo[candidato] > mayor:
        mayor = conteo[candidato]

ganadores = []

for candidato in candidatos:
    if conteo[candidato] == mayor:
        ganadores.append(candidato)

print(conteo)

print("-" * 50)

if len(ganadores) == 1:
    print("Ganador:", ganadores[0])
else:
    print("Empate entre:", ganadores)
```
