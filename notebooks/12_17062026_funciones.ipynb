```python
#1
def es_par(numero: int) -> bool:
    if numero % 2 == 0:
        return True
    return False


print(es_par(10))
print(es_par(5))
```

```python
#2
def mayor(numero_1: int | float, numero_2: int | float) -> int | float:
    if numero_1 > numero_2:
        return numero_1
    return numero_2

print(mayor(5, 8))
print(mayor(10, 8))
```

```python
#3
def promedio(lista: list[int | float]) -> float:

    suma = 0

    for numero in lista:
        suma += numero

    return suma / len(lista)


print(promedio([10, 20, 30, 40]))
print(promedio([5, 8, 9]))
```

```python
#4
def contar_vocales(texto: str) -> int:

    vocales = "aeiou"
    contador = 0

    for letra in texto.lower():

        if letra in vocales:
            contador += 1

    return contador


print(contar_vocales("Python"))
print(contar_vocales("Hola Mundo"))
```
