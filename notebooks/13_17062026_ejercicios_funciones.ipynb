```python
# Ejercicio 1

def analizar_ventas(
    ventas: list[int | float],
    meta: int | float | None = None
) -> dict[str, int | float | bool] | str:

    if not ventas:
        return "No hay ventas"

    for venta in ventas:
        if venta < 0:
            return f"Hay ventas negativas: {venta}"

    total_vendido = sum(ventas)
    promedio = total_vendido / len(ventas)
    venta_mas_alta = max(ventas)
    venta_mas_baja = min(ventas)

    total_superior_promedio = 0
    total_ventas_inferior_100 = 0

    for venta in ventas:
        if venta > promedio:
            total_superior_promedio += 1

        if venta < 100:
            total_ventas_inferior_100 += 1

    resultado = {
        "total_vendido": total_vendido,
        "promedio": promedio,
        "venta_mas_alta": venta_mas_alta,
        "venta_mas_baja": venta_mas_baja,
        "total de ventas superior al promedio": total_superior_promedio,
        "total de ventas inferior a 100": total_ventas_inferior_100
    }

    if meta is not None:
        resultado["meta"] = meta
        resultado["meta_alcanzada"] = total_vendido >= meta

    return resultado
```

```python
# Ejercicio 2

def validar_password(password: str) -> tuple[bool, list[str]]:

    simbolos_requeridos = ("@", "#", "$", "%")

    errores = []

    if len(password) < 8:
        errores.append("El password tiene menos de 8 caracteres")

    contiene_minuscula = False
    contiene_mayuscula = False
    contiene_numero = False
    contiene_simbolos_requeridos = False

    for caracter in password:
        if caracter.islower():
            contiene_minuscula = True
            break

    for caracter in password:
        if caracter.isupper():
            contiene_mayuscula = True
            break

    for caracter in password:
        if caracter.isnumeric():
            contiene_numero = True
            break

    for caracter in password:
        if caracter in simbolos_requeridos:
            contiene_simbolos_requeridos = True
            break

    if not contiene_minuscula:
        errores.append("El password no contiene letra minúscula")

    if not contiene_mayuscula:
        errores.append("El password no contiene letra mayúscula")

    if not contiene_numero:
        errores.append("El password no contiene carácter numérico")

    if not contiene_simbolos_requeridos:
        errores.append("El password no contiene símbolos requeridos")

    if len(password) != len(password.strip()):
        errores.append("El password contiene espacios en blanco")

    es_valida = False if errores else True

    return (es_valida, errores)

```

```python
# Ejercicio 3

def registrar_estudiante(
    estudiantes: dict[str, dict],
    nombre: str,
    notas: list[int | float]
) -> str:

    # Validar nombre
    if not nombre:
        return "Nombre inválido"

    # Validar nombre duplicado
    if nombre in estudiantes:
        return "El nombre ya existe, no se admiten duplicados"

    # Validar mínimo de notas
    if len(notas) < 3:
        return "El estudiante debe tener como mínimo tres notas"

    # Validar notas
    for nota in notas:

        if isinstance(nota, str):
            return "Hay notas no numéricas"

        if nota < 0 or nota > 20:
            return "Hay notas con valores menores a 0 o mayores a 20"

    # Calcular promedio
    promedio = round(sum(notas) / len(notas), 2)

    # Determinar estado
    if promedio >= 11:
        estado = "Aprobado"
    else:
        estado = "Desaprobado"

    # Registrar estudiante
    estudiantes[nombre] = {
        "estado": estado,
        "promedio": promedio,
        "notas": notas
    }

    return "El registro fue exitoso"


# Datos de prueba

estudiantes = {}

print(registrar_estudiante(
    estudiantes,
    "Ana",
    [18, 15, 17]
))

print(registrar_estudiante(
    estudiantes,
    "Luis",
    [10, 12, 9]
))

print(registrar_estudiante(
    estudiantes,
    "Carla",
    [8, 7, 10]
))

print()

print(estudiantes)
```

```python
# Ejercicio 4

def agrupar_por_longitud(palabras: list[str]) -> dict[int, list[str]]:

    palabras_limpias = set(
        palabra.lower()
        for palabra in palabras
        if isinstance(palabra, str)
    )

    print(palabras_limpias)

    grupos = {}

    for palabra_limpia in palabras_limpias:
        longitud = len(palabra_limpia)
        grupos.setdefault(longitud, []).append(palabra_limpia)

    print(grupos)

    for llave in grupos:
        grupos[llave].sort()

    print(grupos)

    return grupos
```

```python
# Ejercicio 5

def actualizar_inventario(
    inventario: dict[str, int],
    producto: str,
    cantidad: int,
    operacion: str
) -> dict[str, str | int]:

    operaciones = ("entrada", "salida")
    operacion = operacion.lower()

    if operacion not in operaciones:
        return {
            "mensaje": f"Error, tipo de operación no soportada: {operacion}",
            "stock_actual": f"{inventario[producto]}"
        }

    if not isinstance(cantidad, int):
        return {
            "mensaje": f"Error, tipo de dato no adecuado, esperaba un entero y consiguió un {type(cantidad)}",
            "stock_actual": f"{inventario[producto]}"
        }

    if cantidad <= 0:
        return {
            "mensaje": "Error, la cantidad es negativa o es cero",
            "stock_actual": f"{inventario[producto]}"
        }

    if operacion == "entrada":
        if producto not in inventario:
            inventario[producto] = cantidad
        else:
            inventario[producto] += cantidad

    if operacion == "salida" and producto in inventario and inventario[producto] >= cantidad:
        inventario[producto] -= cantidad

    return {
        "mensaje": "Operacion exitosa",
        "stock_actual": f"{inventario[producto]}"
    }
```

```python

# Ejercicio 6

productos = {
    "Hamburguesa": 15,
    "Pizza": 25,
    "Gaseosa": 5
}

cantidades = {
    "Hamburguesa": 2,
    "Gaseosa": 3,
    "Helado": 1
}

# calcular_pedido(productos, cantidades, descuento, delivery)
```

```python
def calcular_pedido(
    productos: dict[str, int | float],
    cantidades: dict[str, int],
    descuento: int | float = 0,
    delivery: int | float = 0
) -> dict:

    if descuento < 0 or descuento > 100:
        return {
            "mensaje": "Error, el descuento debe estar entre 0 y 100"
        }

    subtotal = 0
    no_encontrados = []

    for producto, cantidad in cantidades.items():

        if not isinstance(cantidad, int) or cantidad <= 0:
            return {
                "mensaje": f"Cantidad inválida para {producto}"
            }

        if producto not in productos:
            no_encontrados.append(producto)
            continue

        subtotal += productos[producto] * cantidad

    monto_descuento = subtotal * descuento / 100
    total = subtotal - monto_descuento + delivery

    return {
        "subtotal": subtotal,
        "descuento": monto_descuento,
        "delivery": delivery,
        "total": total,
        "no_encontrados": no_encontrados
    }
```

```python
#Ejercicio 7
empleados = {
    "Ana": [80, 90, 75],
    "Carlos": [60, 70, 65],
    "Lucía": [95, 85, 90],
    "Pedro": [70, 70, 70]
}

# evaluar_empleados(empleados, minimo)

```

```python
def evaluar_empleados(
    empleados: dict[str, list[int]],
    minimo: int = 70
) -> list:

    resultado = []

    for nombre, puntajes in empleados.items():

        promedio = sum(puntajes) / len(puntajes)

        if promedio >= minimo:

            if promedio >= 90:
                clasificacion = "Excelente"

            elif promedio >= 80:
                clasificacion = "Bueno"

            elif promedio >= 70:
                clasificacion = "Regular"

            else:
                clasificacion = "Deficiente"

            resultado.append(
                (nombre, promedio, clasificacion)
            )

    resultado.sort(
        key=lambda empleado: (-empleado[1], empleado[0])
    )

    return resultado
```

```python
# Caso del enunciado

resultado = evaluar_empleados(empleados)

print(resultado)
```

```python
def comprimir_texto(texto: str) -> str:

    if texto == "":
        return ""

    resultado = []

    caracter_actual = texto[0]
    contador = 1

    for caracter in texto[1:]:

        if caracter == caracter_actual:
            contador += 1

        else:
            resultado.append(f"{caracter_actual}{contador}")
            caracter_actual = caracter
            contador = 1

    resultado.append(f"{caracter_actual}{contador}")

    return "".join(resultado)
```

```python
def descomprimir_texto(texto: str) -> str:

    resultado = []

    i = 0

    while i < len(texto):

        if not texto[i].isalpha():
            return "Error: formato inválido."

        letra = texto[i]
        i += 1

        numero = ""

        while i < len(texto) and texto[i].isdigit():
            numero += texto[i]
            i += 1

        if numero == "":
            return "Error: formato inválido."

        resultado.append(letra * int(numero))

    return "".join(resultado)
```

```python
#Ejercicio 9
cuentas = {
    "001": {
        "titular": "Ana",
        "saldo": 1500
    },
    "002": {
        "titular": "Luis",
        "saldo": 800
    }
}

# transferir(cuentas, origen, destino, monto)
```

```python
def transferir(
    cuentas: dict,
    origen: str,
    destino: str,
    monto: int | float
) -> dict:

    if origen not in cuentas:
        return {
            "mensaje": "La cuenta de origen no existe."
        }

    if destino not in cuentas:
        return {
            "mensaje": "La cuenta de destino no existe."
        }

    if origen == destino:
        return {
            "mensaje": "Las cuentas deben ser diferentes."
        }

    if not isinstance(monto, (int, float)):
        return {
            "mensaje": "El monto debe ser numérico."
        }

    if monto <= 0:
        return {
            "mensaje": "El monto debe ser mayor que cero."
        }

    if cuentas[origen]["saldo"] < monto:
        return {
            "mensaje": "Saldo insuficiente."
        }

    cuentas[origen]["saldo"] -= monto
    cuentas[destino]["saldo"] += monto

    return {
        "estado": "Transferencia exitosa",
        "origen": cuentas[origen]["titular"],
        "destino": cuentas[destino]["titular"],
        "monto": monto,
        "saldo_restante": cuentas[origen]["saldo"]
    }
```

```python
# Caso del enunciado

resultado = transferir(
    cuentas,
    "001",
    "002",
    300
)

print(resultado)
print(cuentas)
```

```python
# Saldo insuficiente

resultado = transferir(
    cuentas,
    "001",
    "002",
    5000
)

print(resultado)
```

```python
# Cuenta inexistente

resultado = transferir(
    cuentas,
    "003",
    "002",
    100
)

print(resultado)
```

```python
# Mismo origen y destino

resultado = transferir(
    cuentas,
    "001",
    "001",
    100
)

print(resultado)
```

```python
# Monto inválido

resultado = transferir(
    cuentas,
    "001",
    "002",
    -50
)

print(resultado)
```
