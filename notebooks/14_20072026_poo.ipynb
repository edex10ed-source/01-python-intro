```python
class Cuenta:

    def __init__(self, saldo: float):
        self.saldo = saldo

    def depositar(self, monto: float):
        self.saldo += monto


# Crear un objeto
cuenta = Cuenta(1000)

# Mostrar saldo inicial
print("Saldo inicial:", cuenta.saldo)

# Depositar dinero
cuenta.depositar(1000)

# Mostrar saldo final
print("Saldo después del depósito:", cuenta.saldo)
```

```python
#Encapsulamiento
class Cuenta:

    def __init__(self, saldo: float):
        self.__saldo = saldo      # Atributo encapsulado (privado)

    def depositar(self, monto: float):
        self.__saldo += monto

    def retirar(self, monto: float):
        if monto > self.__saldo:
            return "Error: saldo insuficiente"
        self.__saldo -= monto

    def mostrar_saldo(self):
        return self.__saldo
```

```python
# Herencia 
class Animal:

    def __init__(self, es_domestico, num_patas, nombre):
        self.es_domestico = es_domestico
        self.num_patas = num_patas
        self.nombre = nombre

    def comer(self):
        print("Estoy comiendo")

    def beber(self):
        print("Estoy bebiendo agua")

    def es_adoptable(self):
        return self.es_domestico


class Perro(Animal):

    def __init__(self, es_domestico, num_patas, nombre):
        super().__init__(es_domestico, num_patas, nombre)


perro = Perro(True, 4, "Firulais")

print(perro.nombre)
print(perro.num_patas)

perro.comer()
perro.beber()

print(perro.es_adoptable())
```
