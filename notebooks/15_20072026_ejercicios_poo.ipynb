```python
class Persona:

    def __init__(self, nombre: str, edad: int):
        self.nombre = nombre
        self.edad = edad

    def saludar(self):
        print(f"Hola, soy {self.nombre} y tengo {self.edad} años.")


# Crear dos objetos
persona1 = Persona("Ana", 20)
persona2 = Persona("Luis", 25)

# Llamar al método
persona1.saludar()
persona2.saludar()
```

```python
class Rectangulo:

    def __init__(self, base: float, altura: float):
        self.base = base
        self.altura = altura

    def area(self):
        return self.base * self.altura

    def perimetro(self):
        return 2 * (self.base + self.altura)

    def __str__(self):
        return (
            f"Rectángulo\n"
            f"Base: {self.base}\n"
            f"Altura: {self.altura}\n"
            f"Área: {self.area()}\n"
            f"Perímetro: {self.perimetro()}"
        )


# Crear un objeto
mi_rectangulo = Rectangulo(5, 3)

# Mostrar información
print(mi_rectangulo)
    
```

```python
class Coche:

    total_coches = 0

    def __init__(self, marca: str, modelo: str):
        self.marca = marca
        self.modelo = modelo

        self.__incrementar_contador()

    def __incrementar_contador(self):
        Coche.total_coches = Coche.total_coches + 1


coche_1 = Coche("Toyota", "123")
coche_2 = Coche("Ford", "234")
coche_3 = Coche("Coche", "567")

print(coche_1.total_coches)
print(Coche.total_coches)
```
