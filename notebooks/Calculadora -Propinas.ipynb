```python
# Pedir los datos
total_cuenta = float(input("Ingrese el total de la cuenta: S/ "))
porcentaje_propina = float(input("Ingrese el porcentaje de propina (%): "))

# Calcular la propina
monto_propina = total_cuenta * porcentaje_propina / 100
total_pagar = total_cuenta + monto_propina

# Mostrar resultados
print(f"Total de la cuenta : S/ {total_cuenta}")
print(f"Porcentaje         : {porcentaje_propina}%")
print(f"Propina            : S/ {monto_propina}")
print(f"Total a pagar      : S/ {total_pagar}")
```
